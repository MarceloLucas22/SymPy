# Trabalhando com Plotes em SymPy 

### Plote de uma unica variável 

```python
'''
descrição : plotando um gráfico simples da
função f(x) = x² em [-10,10]

'''

from sympy import symbols 
from sympy.plotting import plot 

# criando a função a ser plotada :

x = symbols('x', real = True )

fun = x**2 

# plotando essa função : 

plot(fun,(x,-10,10))

```

### plotando várias funções em uma mesma janela 

```python
'''
descrição : plote das funções x^2,x^3,x^4 em [-1,1].
programador : Marcelo Lucas Almeida 
'''
# importando os modulos : 
from sympy import symbols 
from sympy.plotting import plot    # importa o sub-modulo de plotagem

# criando a variável simbolica :
x = symbols("x", real = True)
# criando as funções :
fun1 = x**2
fun2 = x**3 
fun3 = x**4
# criando o plote  
plot(fun1,fun2,fun3,(x,-1,1))  
```

### grafico de uma variável personalizado 

```python
'''
descrição :  vamos construir o grafico da função 
f(x) = sin(x) em [-pi,pi]
=====================================
programador : Marcelo L S Almeida
'''

from sympy.plotting import plot 
from sympy import symbols, sin , pi 

# criando o simbolo x :
x = symbols("x", real = True)

f_x = sin(x)       # função f(x) = sin(x)

# personalizando o grafico :

plot(f_x,(x,-pi,pi),xlabel='eixo x', ylabel='eixo y',
title='plote da função sen(x)', line_color = 'g', legend = True)
```

### plotando dois graficos usando extend() 

```python
'''
descrição : plotar os graficos de f(x) = sin(2x) e g(x) = cos(2x)
em [-pi,pi]
=================================================================
programador : Marcelo L S Almeida
'''
from sympy.plotting import plot 
from sympy import symbols, sin , pi , cos 
# criando o simbolo x :
x = symbols("x", real = True)
# criando as funçções :
f_x = sin(2*x)
g_x = cos(2*x)
# criando os gráficos :
p1 = plot(f_x,(x,-pi,pi),xlabel='eixo x', ylabel='eixo y',
title='plote de sin(2x) e cos(2x)',legend = True,
line_color = 'r', show= False)
p2 = plot(g_x,(x,-pi,pi),legend = True,
line_color = 'g', show= False)
p1.extend(p2) # adiciona o gráfico p2 em p1 
p1.show()
```

### plotando quatro termos da serie de taylor da exponencial 

```python
'''
descrição : vamos plotar o gráfico da função f(x) = e^x e 
o seus quatros termos da série de taylor em x0 = 0 em [-1,1] 
=================================================================
programador : Marcelo L S Almeida
'''
from sympy.plotting import plot 
from sympy import symbols , exp
import math as mt  
# criando os simbolos :
x = symbols("x", real = True)
k = symbols("k", integer = True)
# criando as funções :
funcao = exp(x)    # função exponencial    
expr1 = (x**k)  # função que representa o numero da serie de taylor de exp(x)
expr2 = k       # denominador da expressão da serie de taylor de exp(x)    
# criando uma lista : 
lista = [0,1,2,3]   # termos da série 
lista_cor = ['r','g','k','c'] # lista de cores    
# criando o plote que receberá os termos da serie :
p = plot(funcao,(x,-1,1),xlabel='eixo x', ylabel='eixo y',
title='plote de e^x e alguns termos da serie de taylor',
line_color = 'b',legend = True,show=False)    
# criando um laço para plotar :
valor = 0      # acumula as funções  
for i in range(len(lista)):
        valor += expr1.subs({k:lista[i]})/mt.factorial(expr2.subs({k:lista[i]}))
    
        p1 = plot(valor,(x,-1,1),line_color = lista_cor[i],
            legend = True, show=False)
    
        p.extend(p1)
        
p.show()
```

### Plote de funções parametricas 

```python
'''
descrição : plote da curva parametrica de equações : (sin(t),sin(t)cos(t))

com t em [0,2pi]
===================================================================================
programador : Marcelo Almeida 
'''

from sympy.plotting import plot_parametric
from sympy import symbols, pi, sin , cos   

# criando as funções :
t = symbols('t', real = True )   # parametro 
x_t = sin(t)                     # x(t)
y_t = sin(t)*cos(t)              # y(t)
# plote da curva parametrizada 
plot_parametric(x_t,y_t,(t,0,2*pi), xlabel='x(t)',ylabel='y(t)',
               title = 'Plote da curva parametrica (sin(t),sin(t)cos(t))')
```

### Curva Borboleta 

```python
'''
descrição : plote da nó de trevo : (sin(t) + 2sint(2t),cos(t) - 2cos(2t))

com t em [0,2pi]
===================================================================================
programador : Marcelo Almeida 
'''
from sympy.plotting import plot_parametric
from sympy import symbols, pi, sin , cos   

# criando as funções :
t = symbols('t', real = True)
x_t = sin(t) + 2*sin(2*t)     # x(t)
y_t = cos(t) - 2*cos(2*t)     # y(t)

# plote dessa função :

plot_parametric(x_t,y_t,(t,0,2*pi), xlabel='x(t)', ylabel='y(t)',
               title='plote de  (sin(t) + 2sint(2t),cos(t) - 2cos(2t))')
```

### Curva Borboleta

```python
'''
descrição : plote da curva borboleta em t = [0,100]
===================================================================================
programador : Marcelo Almeida 
'''
from sympy.plotting import plot_parametric
from sympy import symbols, pi, sin , cos  , exp 

# criando as funções : 
t = symbols('t', real = True )
x_t = sin(t)*(exp(cos(t)) - 2*cos(4*t) - sin(t/12)**5)
y_t = cos(t)*(exp(cos(t)) - 2*cos(4*t) - sin(t/12)**5)

# plote da função parametrica :

plot_parametric(x_t,y_t,(t,0,100), title='plote da curva borboleta', line_color = 'r')
```

### Plotagem de funções implicitas 

```python
'''
descrição : plote da equação implicita dada por : 
x² + y² = 3 
com x = [-3,3] e y = [-3,3]
==========================================

programador : Marcelo Lucas Almeida 
'''
from sympy.plotting import plot_implicit 
from sympy import symbols , Eq 

# definindo os simbolos : 
x , y = symbols('x,y', real = True)

# criando a equação implicita :
fun_implicit = Eq(x**2 + y**2 , 3)

# plote dessa equação : 
plot_implicit(fun_implicit,(x,-3,3),(y,-3,3), xlabel='eixo x',
             ylabel='eixo y',title='plote da equação x² + y² = 3')
```

```python
'''
descrição : plote da inequação dada por y > x² 
==============================================
programador : Marcelo Lucas Almeida
'''
from sympy.plotting import plot_implicit
from sympy import symbols 

# criando os simbolos : 
x , y = symbols('x,y', real = True)

# plotando a inequação : 
plot_implicit(y>x**2, xlabel='eixo x', ylabel='eixo y', title='plote da inequação y > x²')
```

```python
'''
descrição : plote da inequação dada por y > x e y > -x
=======================================================
programador : Marcelo Lucas Almeida 
'''
from sympy.plotting import plot_implicit
from sympy import symbols , And 

# criando os simbolos : 
x , y = symbols('x,y', real = True)

# plotando a inequação : 
plot_implicit(And(y > x , y > -x), title='plote da inequação y > x e y > -x')
```

### Plote 3D


```python
'''
descrição : plote da função f(x,y) = cos(y)sin(x) em [-6,6]x[-6,6]
==================================================================
programador : Marcelo Lucas Almeida

'''
from sympy import symbols , sin , exp , cos 
from sympy.plotting import plot3d
x , y = symbols('x,y',real = True)

# criando a função f(x,y)
fun_xy = cos(y)*sin(x)

# criando o plote da função 
plot3d(fun_xy,(x,-6,6),(y,-6,6))
```

```python
'''
descrição : PLote da função f(x,y) = (4x² + y²)exp(-x² - y²) em [-2,2]x[-2,2]
=============================================================
Programador : Marcelo Lucas Almeida
'''
from sympy.plotting import plot3d
from sympy import symbols, exp 
import math as mt 

# criando os simbolos : 
x,y = symbols('x,y', real = True )

# função 
f_xy = (4*x**2 + y**2)*exp(-x**2 - y**2)

# plote da função : 
plot3d(f_xy,(x,-2,2),(y,-2,2), xlabel='eixo x', ylabel='eixo y', zlabel='f(x,y)')
```

```python
'''
descrição : plote dos paraboloides f(x,y)= x² + y² e g(x,y) = -(x² + y²) em [-10,10]x[-10,10]
===============================================================================================
programador : Marcelo Lucas Alemida 
'''
from sympy.plotting import plot3d
from sympy import symbols 

# criando as funções 
f_xy = x**2 + y**2 
g_xy = -x**2 - y**2 

# plote dessas duas funções :
plot3d(f_xy,g_xy,(x,-10,10),(y,-10,10),xlabel='eixo x ',ylabel='eixo y')
```

### Cruvas parametricas 3d

```python
'''
descrição : vamos plotar a curva parametrica dada pelas equações : 

(\sqrt(t)cos(4t), \sqrt(t)sin(4t), \sqrt(t))

com t variando no intervalor de [0,2pi]
===================================================================
programador : Marcelo Lucas Almeida
'''
from sympy.plotting import plot3d_parametric_line
from sympy import symbols, sqrt , cos , sin , pi  
# criando o nosso parametro :
t = symbols('t', real = True)
# criando as funções de t : 
x_t = sqrt(t)*cos(4*t)
y_t = sqrt(t)*sin(4*t)
z_t = sqrt(t)
# plote da curva parametrica : 
plot3d_parametric_line(x_t,y_t,z_t,(t,0,2*pi))
```

```python
'''
descrição : vamos plotar a curva parametrica dada pelas equações : 

(e^{-|x|/10}sin(5|x|), e^{-|x}/10}cos(5|t|), t)

com t variando no intervalor de [-5,5]
===================================================================
programador : Marcelo Lucas Almeida
'''
from sympy.plotting import plot3d_parametric_line
from sympy import symbols, exp , sin , cos 

# criando o parametro :
t = symbols('t',real = True)
# criando as funções de t :
x_t = exp(-abs(t)/10)*sin(5*abs(t))
y_t = exp(-abs(t)/10)*cos(5*abs(t))
z_t = t 
#plote da equação parametrica :
plot3d_parametric_line(x_t,y_t,z_t,(t,-5,5))
```

### Superficies Paramtrizadas 

```python
'''
descrição : Plote da superficie paramtrica de equação : 

(2sin(u)cos(v), 2sin(v)sin(u), 2cos(v))
===============================================
programador : Marcelo Almeida
'''
from sympy.plotting import plot3d_parametric_surface
from sympy import symbols , sin , cos , pi

#criando os parametros u e v : 
u , v = symbols("u,v", real = True)
# funções parametricas da superficie :
x_uv = 2*sin(v)*cos(u)
y_uv = 2*sin(v)*sin(u)
z_uv = 2*cos(v)

# plote da superficie parametrizada : 
plot3d_parametric_surface(x_uv,y_uv,z_uv,(u,0,2*pi),(v,0,pi),title='superficie parametrica',axis ='equal')
```

```python
'''
descrição : Plote da superficie paramtrica de equação : 

(cos(u) - vsin(u), sin(u) + vcos(u),v)
===============================================
programador : Marcelo Almeida
'''
from sympy.plotting import plot3d_parametric_surface
from sympy import symbols , sin , cos , pi

#criando os parametros u e v : 
u , v = symbols("u,v", real = True)
# funções parametricas da superficie :
x_uv = cos(u) - v*sin(u)
y_uv = sin(u) + v*cos(u)
z_uv = v

# plote da superficie parametrizada : 
plot3d_parametric_surface(x_uv,y_uv,z_uv,(u,0,2*pi),(v,-2,2),title='Hiperboloide de uma folha')
```
